### Exercice 61


```python
import matplotlib.pyplot as plt
from numpy import *
from scipy.optimize import *

u = lambda x : log(x)-5+x
v = lambda x: (x-2)**2 - log(x)
g = lambda x: exp(-x)-x
f = lambda x: x**3 + 4*x**2 -10

func_ar = [u,v,g,f]

i = linspace(1,2,20000)
plt.grid(color='r', linestyle='-', linewidth=0.1)


for func in func_ar:
    plt.plot(i,func(i))
    
print(fsolve(u,1.4))
print(fsolve(v,1))
```

### Exercices 62


```python
import matplotlib.pyplot as plt
from numpy import *
from scipy.optimize import *
u = lambda x: x*exp(x)
du = lambda x: exp(x) + x*exp(x)

x0 = 20

dx = -(u(x0))/(du(x0))

eps = 0.000000000001

while abs(dx) > eps:
    dx = -(u(x0))/(du(x0))
    x0 = x0 + dx

print(x0)
i = linspace(-1,1,1000)

plt.grid(color='r', linestyle='-', linewidth=0.1)
print(fsolve(u, 20))
plt.plot(i,u(i))
```

### Exercices 63


```python
import matplotlib.pyplot as plt
from numpy import *

u = lambda x : x**2-4*(x-1)
func_step = lambda x : 2 * sqrt(x-1)

eps = 1e-5
x0 = 2.5

while abs(u(x0)) > eps:
    x0 = func_step(x0)
    print(x0)
print(x0)
```

### Exercices 68


```python
import matplolib.pyplot as plt
from numpy import *
from numpy.linalf import *

def f(x,y):
    return [x**2 - y , 4*x**2 +9 *y**2 -8*x -32]

def Jf(x,y):
    return [[2*x,-1],[8*x-8,18*y]]

x0 = 1.4
y0 = 2
eps = 1e-3
dxdy = solve(Jf(x0,y0), f(x0,y0))

while abs(dxdy[0])> eps and abs(dxdy[1])<eps:
    dxdy = solve(Jf(x0,y0), f(x0,y0))
    x0 = x0 +dxdy[0]
    y0 = y0 + dxdy[1]

print(x0,y0)
```
